```python
file = "poverty_census_tract"
poverty_csv = pd.read_csv(file)
poverty_csv.head()
```

```python
merged = pd.merge(poverty_csv, crime_csv, on= 'clean_tract', how= 'inner')
merged
```

```python
merged.info()
```

```python
merged_df = merged[['clean_tract', 'B17020_001E','B17020_calc_pctPovE', 'B17020_calc_pctChildPovE', 'Offense', 'Method', 'Latitude', 'Longitude', 'Start_Date', 'End_Date','CCN']]
merged_df
```

```python
number_of_crimes = merged_df.groupby('clean_tract').count()['Offense']
```

```python
offense_type= merged_df.groupby('Offense').count()['clean_tract']
offense_type
```

```python
population = merged_df.groupby('clean_tract').mean()['B17020_001E']
```

```python
pop_pct = merged_df.groupby('clean_tract').mean()['B17020_calc_pctPovE']
```

```python
merged_df['B17020_calc_pctChildPovE'].fillna(merged_df['B17020_calc_pctChildPovE'].mean(), inplace = True)
childpop_pct = merged_df.groupby('clean_tract').mean()['B17020_calc_pctChildPovE']
childpop_pct
```

```python
lat = merged_df.groupby('clean_tract').mean()['Latitude']
lat
```

```python
long = merged_df.groupby('clean_tract').mean()['Longitude']
long
```

```python
summary_table = pd.DataFrame({"Census": merged_df.clean_tract.unique(),
                              "Offense": offense_type,
                              "Number of Crimes": number_of_crimes,
                              "Total Population": population,
                              "Percentage of Population in Poverty": pop_pct,
                              "Percentage of Children in Poverty": childpop_pct,
                              "Latitude": lat,
                              "Longitude": long})
```

```python
summary_table = summary_table.reset_index(drop= True)
```
