```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = "../Resources/ACS_2017_Poverty_Status_Variables_Tract.csv"
poverty_csv = pd.read_csv(file)
poverty_csv.head()
```

```python
poverty_csv.info()
```

```python
poverty_df = pd.DataFrame(poverty_csv, columns=['NAME','B17020_calc_pctPovE', 'B17020_calc_pctChildPovE'])
poverty_df
```

```python
poverty_df.info()
```

```python
#Rename Columns
poverty_df.rename(columns = {'NAME': 'Census Tract','B17020_calc_pctPovE':'Percentage of Poverty', 'B17020_calc_pctChildPovE':'Percentage of Poverty in Children'}, inplace = True)
poverty_df
```

```python
poverty_df1 = poverty_df.fillna(0)
poverty_df1
```

```python
percentage_of_povt10 = poverty_df1.sort_values('Percentage of Poverty', ascending=False)
percentage_of_povt10.set_index("Census Tract")
percentage_of_povt10.head(10)
```

```python
#Make a barchart with the Top 10 for Perecentage of Poverty
x_axis = np.arange(len(percentage_of_povt10.head(10)))
bar =percentage_of_povt10.head(10)["Percentage of Poverty"]
tick_locations = [value for value in x_axis]
bar.plot(kind="bar", align = 'center')
plt.xlabel("Census Tract")
plt.xticks(tick_locations, percentage_of_povt10['Census Tract'], rotation=45,)
plt.ylabel("Percentage of Poverty")
plt.title("Top 10 Areas of Highest Percentage of Poverty")
plt.show()
```

```python
#Barchart with Top 10 Percentage of Child Poverty
percentage_of_child_pov10 = poverty_df1.sort_values('Percentage of Poverty in Children', ascending=False)
percentage_of_child_pov10.set_index("Census Tract")
percentage_of_child_pov10.head(10)
```

```python
x_axis = np.arange(len(percentage_of_child_pov10.head(10)))
bar =percentage_of_child_pov10.head(10)["Percentage of Poverty in Children"]
tick_locations = [value for value in x_axis]
bar.plot(kind="bar", align = 'center')
plt.xlabel("Census Tract")
plt.xticks(tick_locations, percentage_of_child_pov10['Census Tract'], rotation=45,)
plt.ylabel("Percentage of Poverty in Children")
plt.title("Top 10 Areas of Children in Poverty")
plt.show()
```

```python
#Barchart with Lowest 10 Poverty
bottom_ten = poverty_df1.sort_values('Percentage of Poverty', ascending=True)
bottom_ten.set_index("Census Tract")
bottom_ten.head(10)
```

```python
x_axis = np.arange(len(bottom_ten.head(10)))
bar =bottom_ten.head(10)["Percentage of Poverty"]
tick_locations = [value for value in x_axis]
bar.plot(kind="bar", align = 'center')
plt.xlabel("Census Tract")
plt.xticks(tick_locations, bottom_ten['Census Tract'], rotation=45,)
plt.ylabel("Percentage of Poverty")
plt.title("Bottom 10 Areas of Poverty Percentage") 
plt.show()
```

```python
#Barchart with Lowest 10 Children in Poverty
bottom_ten_children = poverty_df1.sort_values('Percentage of Poverty in Children', ascending=True)
bottom_ten_children.set_index("Census Tract")
bottom_ten_children.head(10)
```

```python
x_axis = np.arange(len(bottom_ten_children.head(10)))
bar =bottom_ten_children.head(10)["Percentage of Poverty in Children"]
tick_locations = [value for value in x_axis]
bar.plot(kind="bar", align = 'center')
plt.xlabel("Census Tract")
plt.xticks(tick_locations, bottom_ten_children['Census Tract'], rotation=45,)
plt.ylabel("Percentage of Poverty")
plt.title("Bottom 10 Areas of Children in Poverty Percentage") 
plt.show()
```
