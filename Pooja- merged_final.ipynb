```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = "Resources/merged_final.csv"
data = pd.read_csv(file)
data
```

```python
sortedinfo= data.sort_values('Pct_Pop_Below_Pov', ascending=True)
sortedinfo
```

```python
gbdata = data.groupby(['Census_Tract','Offense']).count()
gbdata
```

```python
gbdata = gbdata.reset_index()
gbdata
```

```python
data201 = gbdata[gbdata['Census_Tract']==201]
data201
```

```python
#Take the Area with the Highest Amount of Poverty and bargraph the types of crime that occur
x_axis = data201['Offense']
bar =data201["Pct_Pop_Below_Pov"]
tick_locations = [value for value in x_axis]
bar.plot(kind="bar", align = 'center')
plt.xlabel("Type of Offense")
plt.xticks(np.arange(len(tick_locations)), tick_locations, rotation = 90)
plt.ylabel("Amount")
plt.title("Census 201: 100% Poverty Rate")
plt.show()
```

```python
#Take the area with the lowest amount of poverty and bargraph what types of crimes
data6202 = gbdata[gbdata['Census_Tract']==6202]
data6202
```

```python
x_axis = data6202[('Offense')]
bar =data6202["Pct_Pop_Below_Pov"]
tick_locations = [value for value in x_axis]
bar.plot(kind="bar", align = 'center')
plt.xlabel("Type of Offense")
plt.xticks(np.arange(len(tick_locations)), tick_locations, rotation = 90)
plt.ylabel("Amount")
plt.title("Census 6202: 0% Poverty Rate")
plt.show()
```
